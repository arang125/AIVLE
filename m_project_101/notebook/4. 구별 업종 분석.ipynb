#### [4.2.2] 원하는 데이터 형태로 저장하기

* head, tail 등을 살펴본 결과를 토대로 header 값을 적절히 지정하여 데이터를 다시 불러오고 필요한 컬럼만 저장하세요

```python
#read_csv 재실행. 단, header값을 적절히 지정
seoul_business.columns = [x+" "+y if x!=y else x for x,y in zip(seoul_business.iloc[0], seoul_business.iloc[1])]
seoul_business.columns
```

```python
# 다시 불러온 데이터프레임에서 여러분 각자가 사용하고자 하는 컬럼만 가져오세요.

selected_column = ["자치구", "동"] + [column for column in seoul_business.columns if '종사자수' in column]
seoul_business = seoul_business[selected_column]

seoul_business[2:].head(20)
```

```python
# 여러분 각자가 사용하고자 하는 형태로 데이터를 정리합니다. 
# 예를 들어, 동 컬럼의 값이 소계인 항목들만 골라내 보겠습니다. 
# 위와 같이 동 컬럼의 소계를 구하면 자치구 별 값이 되므로 이후에 동 컬럼은 삭제하겠습니다. 결과적으로 자치구 별 값을 얻게 됩니다.
seoul_business = seoul_business.loc[seoul_business['동'] == '소계']
seoul_business = seoul_business.drop('동', axis=1)
seoul_business.head()
```

```python
seoul_business.set_index('자치구', inplace=True)
seoul_business = seoul_business.apply(lambda x: pd.to_numeric(x.str.replace(',', ''), errors='coerce'))
seoul_business.info()
```

```python
# 위에서 여러분 각자가 사용하고자 하는 형태로 정리한 해당 데이터프레임을 csv 파일('df_seoul_business.csv')로 저장하세요.
seoul_business.to_csv('df_seoul_business.csv')
```

# 4.3.데이터 분석하기
+ 데이터의 형태를 살펴보고 다양한 분석기법을 통해 모델링에 적합하도록 정제요소를 선별하세요.

```python
# 시각화 한글폰트 설정
import seaborn as sns
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
sns.set(font="Malgun Gothic",#"NanumGothicCoding", 
        rc={"axes.unicode_minus":False}, # 마이너스 부호 깨짐 현상 해결
        style='darkgrid')        
```

#### [4.3.1] 데이터 분포 알아보기_1
* '자치구' 등 여러 변수를 기준으로 시각화 해보고 인사이트를 도출해보세요.

```python
# 아래에 코드를 작성하고 결과를 확인합니다.
sns.heatmap(seoul_business.corr(), annot=True)
plt.show()
```

```python
# 아래에 코드를 작성하고 결과를 확인합니다.
plt.figure(figsize=(15,8))

plt.subplot(2,2,1)
sns.barplot(x="자치구", y="택시운송업 종사자수", data=seoul_business)
plt.xticks(rotation=45)

plt.subplot(2,2,2)
sns.barplot(x="자치구", y="용달 화물자동차 운송업 종사자수", data=seoul_business)
plt.xticks(rotation=45)

plt.subplot(2,2,3)
sns.barplot(x="자치구", y="커피전문점 종사자수", data=seoul_business)
plt.xticks(rotation=45)

plt.subplot(2,2,4)
sns.barplot(x="자치구", y="기타주점업 종사자수", data=seoul_business)
plt.xticks(rotation=45)

plt.show()
```
