# 1. 데이터 전처리

#### [실습문제1] 열 추가
* 기존 '기준일ID'열에서 날짜 데이터를 추출하여 새로운 열로 추가 합니다.
    - pd.to_datetime() 함수를 사용해 '기준일ID'열을 datetime 형식으로 변경하겠습니다.
    - year, month, day 열로 기존 데이터 프레임에 연월일 정보를 저장합니다.
    - 데이터 프레임 열에 대해 날짜 정보를 얻을 때에는 **dt.year**, **dt.month**, **dt.day** 속성을 사용합니다.

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
train_set['기준일ID'] = pd.to_datetime(train_set['기준일ID'].astype(str))
train_set['year'] = train_set['기준일ID'].apply(lambda d: d.year)
train_set['month'] = train_set['기준일ID'].apply(lambda d: d.month)
train_set['day'] = train_set['기준일ID'].apply(lambda d: d.day)
#train_set['weekday'] = train_set['기준일ID'].apply(lambda d: d.weekday())
train_set.head(1)
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
test_set['기준일ID'] = pd.to_datetime(test_set['기준일ID'].astype(str))
test_set['year'] = test_set['기준일ID'].apply(lambda d: d.year)
test_set['month'] = test_set['기준일ID'].apply(lambda d: d.month)
test_set['day'] = test_set['기준일ID'].apply(lambda d: d.day)
#test_set['weekday'] = test_set['기준일ID'].apply(lambda d: d.weekday())
test_set.head(0)
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
man_columns = [column for column in train_set.columns if '남자' in column]
woman_columns = [column for column in train_set.columns if '여자' in column]

train_set['남녀성비'] = train_set[man_columns].sum(axis=1)/train_set[woman_columns].sum(axis=1)
test_set['남녀성비'] = test_set[man_columns].sum(axis=1)/test_set[woman_columns].sum(axis=1)
train_set[:1]
```

#### [실습문제2] 불필요한 열 제거 및 열 이름 변경
* df_total과 df_test에서 '총생활인구수'외에 불필요한 열을 제거 하도록 하겠습니다.
* 시간 데이터를 남기고, '시간대구분' column 을 'hour'로 rename 하도록 하겠습니다.

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
drop_cols = list(filter(lambda s: '세부터' in s or "이상" in s, test_set.columns))
train_set.drop(drop_cols, axis=1, inplace=True)
test_set.drop(drop_cols, axis=1, inplace=True)
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
train_set.rename(columns={"시간대구분": "hour"}, inplace=True)
test_set.rename(columns={"시간대구분": "hour"}, inplace=True)
```

```python
train_set.head(0)
```

```python
train_set = train_set.groupby(['month', 'day', 'hour']).mean().reset_index().drop(['year'], axis=1)
train_set.head()
```

```python
test_set = test_set.groupby(['month','day', 'hour']).mean().reset_index().drop(['year'], axis=1)
test_set.head()
```

#### [실습문제3] 시프트 & 롤링
* shift()를 활용하여 이동할 행 수를 지정합니다.
* rolling()을 활용하여 추가할 이동평균 수를 지정합니다.

```python
target = "총생활인구수"
```

* 라그 변수 (shift)

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
train_set["총생활인구수_1시간전"] = train_set["총생활인구수"].shift(-1)
train_set["총생활인구수_3시간전"] = train_set["총생활인구수"].shift(-3)
train_set["총생활인구수_24시간전"] = train_set["총생활인구수"].shift(-24)
display(train_set)
test_set["총생활인구수_1시간전"] = test_set["총생활인구수"].shift(-1)
test_set["총생활인구수_3시간전"] = test_set["총생활인구수"].shift(-3)
test_set["총생활인구수_24시간전"] = test_set["총생활인구수"].shift(-24)
```

* 이동 평균 (rolling)

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.

train_set['총생활인구수_3시간전평균'] = train_set["총생활인구수"].shift(1).rolling(3, min_periods=1).mean()
train_set['총생활인구수_6시간전평균'] = train_set["총생활인구수"].shift(1).rolling(6, min_periods=1).mean()
train_set['총생활인구수_12시간전평균'] = train_set["총생활인구수"].shift(1).rolling(12, min_periods=1).mean()
train_set['총생활인구수_24시간전평균'] = train_set["총생활인구수"].shift(1).rolling(24, min_periods=1).mean()
display(train_set)
test_set['총생활인구수_3시간전평균'] = test_set["총생활인구수"].shift(1).rolling(3, min_periods=1).mean()
test_set['총생활인구수_6시간전평균'] = test_set["총생활인구수"].shift(1).rolling(6, min_periods=1).mean()
test_set['총생활인구수_12시간전평균'] = test_set["총생활인구수"].shift(1).rolling(12, min_periods=1).mean()
test_set['총생활인구수_24시간전평균'] = test_set["총생활인구수"].shift(1).rolling(24, min_periods=1).mean()
```

#### [실습문제4] 결측치 처리 및 불필요한 열 제거
* 불필요한 열을 추가로 제거 하겠습니다.(시간 관련 행들)
* 롤링 & 시프트 과정에서 발생한 결측치를 처리하도록 하겠습니다.


```python
train_set.drop('기준일ID', axis=1, inplace=True)
test_set.drop('기준일ID', axis=1, inplace=True)

train_set.dropna(axis=0, inplace=True)
test_set.dropna(axis=0, inplace=True)
```

#### [실습문제5] train, test 데이터 나누기
* df_total과 df_test를 각각 train_x, train_y, test_x, test_y 로 나누어 주세요.
* 인덱스를 리셋 해주세요.

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.

train_x, train_y = train_set.drop(target, axis=1), train_set[target]
test_x, test_y = test_set.drop(target, axis=1), test_set[target]
```
